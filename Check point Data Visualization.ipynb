```python
Title_Dictionary = {
                    "Capt"        :  "Officer",
                    "Col"         :  "Officer",
                    "Major"       :  "Officer",
                    "Dr"          :  "Officer",
                    "Rev"         :  "Officer",
                    "Jonkheer"    :  "Royalty",
                    "Don"         :  "Royalty",
                    "Sir"         :  "Royalty",
                    "Lady"        :  "Royalty",
                    "the Countess":  "Royalty",
                    "Dona"        :  "Royalty",
                    "Mme"         :  "Miss",
                    "Mlle"        :  "Miss",
                    "Miss"        :  "Miss",
                    "Ms"          :  "Mrs",
                    "Mr"          :  "Mrs",
                    "Mrs"         :  "Mrs",
                    "Master"      :  "Master"
                    }
```

```python
#### Expml de chaine de caract
string=" Moen, Mr. Sigurd Hansen"
part_string = string.split('.')[0] ## split ta9sem win tsib point wla virgule wla...
title = part_string.split(',')[1]
print(title)
```

```python
title = title.strip() ## split pour retirer les espace dans la ch de car
grade = Title_Dictionary[title]
print(grade)
```

```python
def recup_titre(x):
  part_string = x.split('.')[0]
  title = part_string.split(',')[1]
  title = title.strip()
  grade = Title_Dictionary[title]
  return grade

df['Title'] = df['Name'].apply(recup_titre)
df
```
